## **Data Preparation**

### Merging Datasets

```python
# We are groupmg the film titles by regions to get the number of regions that will enable us merge it into the primary data set which is title-basics
title_akas_group=title_akas.groupby('title_id')['region'].count().reset_index(name='No_of_regions')
title_akas_group
```

```python
# We are creating a pivot table with film on the row and the crew categories on the columns. This will enable us get the number of crews who participated in the creation of the film
title_principals_pivot=title_principals.pivot_table(index='tconst', columns='category',values='ordering',aggfunc='count', fill_value=0)
title_principals_pivot['total_crew'] = title_principals_pivot.sum(axis=1)
title_principals_pivot
```

We are are merging secondary data sets into the priamry data set which is title_basics. The secondary data to be merged are:
1. Title ratings- This will give us the ratings of the films and the number of votes
2. title_principals_pivot: This will give the number of crews who participated per category
3. bom_movie: This will give us the financial infomation with regards to the films
4. title_aka_groups: This will give the regions in which tha film was filmed.

```python
# Merging title_ratings and title_principal_pivot into the primary dataset title_basics
movies_data= (pd.merge(title_basics,title_ratings,on="tconst",how="left").merge(title_principals_pivot, on='tconst',how="left"))
# Merging bom.movie_gross into the primary data set title_basics
movies_data1= pd.merge(movies_data, bom_movie, left_on=['primary_title', 'start_year'], right_on=['title', 'year'],how="left")

# We wanted to get the genres in one film however there were items in the genre column that were null hence we createf below function to help us
def count_genres(genre_string):
    # Check if the value is NaN
    if isinstance(genre_string, str):
        return len(genre_string.split(','))
    else:
        return 0  # Return 0 for NaN values
# Creating the number of genres per film
movies_data1['genre_counts'] = movies_data1['genres'].apply(count_genres)
# Merging title_akas_group into the primary data set title_basics
movies_data2=pd.merge(movies_data1, title_akas_group, left_on=['tconst'], right_on=['title_id'],how="left")
movies_data2.head()
```

### Removing outliers

```python
#create a function to drop outliers using IQR
def drop_outliers_IQR(movies_data5, column):

   q1=movies_data5[column].quantile(0.25)
   q3=movies_data5[column].quantile(0.75)

   IQR=q3-q1
   movies_data4 = movies_data5[((movies_data5[column]>(q1-1.5*IQR)) | (movies_data5[column]<(q3+1.5*IQR)))]

   return movies_data4

# Find the outliers in the bmi column using the IQR function created
movies_data4=drop_outliers_IQR(movies_data5, 'runtime_minutes')
```

#### Popularity of Genres

After analyzing the data, it becomes evident that multi-genre films, comprising either double or triple genres, tend to garner higher average numbers of votes compared to single-genre films.

```python
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Assign numerical values to genre counts
def assign_genre_category(genre_count):
    if genre_count == 1:
        return 'Single Genre'
    elif genre_count == 2:
        return 'Double Genre'
    elif genre_count == 3:
        return 'Triple Genre'
    else:
        return 'Others'

# Create a new column 'genre_category'
movies_data4['genre_category'] = movies_data4['genre_counts'].apply(assign_genre_category)

# Convert 'numvotes' column to integers
movies_data4['numvotes'] = movies_data4['numvotes'].astype(int)

# Group the data by 'genre_category' and calculate the total number of votes for each category
numvotes_by_genre_category = movies_data4.groupby('genre_category')['numvotes'].sum().reset_index()

# Format the 'numvotes' column to display numbers with commas
numvotes_by_genre_category['numvotes'] = numvotes_by_genre_category['numvotes'].apply(lambda x: '{:,}'.format(x))

# Display the DataFrame
numvotes_by_genre_category
```

```python
# Define colors for each genre
colors = ['blue', 'red', 'green', 'yellow']

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(numvotes_by_genre_category['genre_category'], numvotes_by_genre_category['numvotes'], color=colors)
plt.title('Number of Votes by Genre Category')
plt.xlabel('Genre Category')
plt.ylabel('Number of Votes')
plt.xticks(rotation=45)
plt.show()
```

 ### Popularity of leght of movies in minutes

 The observation based on the provided data is that the number of votes generally increases with the runtime interval of the films. Specifically, films with longer durations, ranging from 101 to 120 minutes, receive the highest number of votes, followed by those with durations between 121 and 140 minutes. Conversely, films with shorter durations, such as those in the 0-20 and 21-40 minute intervals, receive relatively fewer votes. This trend suggests that viewers may prefer films with longer runtimes, as they potentially offer more engaging and substantial content.

```python
# Suppress SettingWithCopyWarning
with warnings.catch_warnings():
    warnings.simplefilter("ignore")

    # Define the runtime intervals
    runtime_intervals = [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100), (101, 120),
                         (121, 140), (141, 160), (161, 180), (181, 200), (201, 100000)]

    # Function to assign runtime intervals
    def assign_runtime_interval(runtime):
        for interval in runtime_intervals:
            if interval[0] <= runtime <= interval[1]:
                return f"{interval[0]}-{interval[1]}"

    # Apply the function to create the new column
    movies_data4['runtime_interval'] = movies_data4['runtime_minutes'].apply(assign_runtime_interval)

# Group the data by runtime interval and calculate the total number of votes for each interval
numvotes_by_runtime = movies_data4.groupby('runtime_interval')['numvotes'].sum().reset_index()

# Define the order of runtime intervals as strings
interval_order = [f"{interval[0]}-{interval[1]}" for interval in runtime_intervals]

# Convert 'runtime_interval' to ordered categorical type with specified order
numvotes_by_runtime['runtime_interval'] = pd.Categorical(numvotes_by_runtime['runtime_interval'], categories=interval_order, ordered=True)

# Sort the DataFrame by 'runtime_interval' in ascending order
numvotes_by_runtime_sorted = numvotes_by_runtime.sort_values(by='runtime_interval')
numvotes_by_runtime_sorted['numvotes'] = numvotes_by_runtime_sorted['numvotes'].apply(lambda x: '{:,}'.format(x))
numvotes_by_runtime_sorted
```

```python
# Convert the 'numvotes' column to string type
numvotes_by_runtime_sorted['numvotes'] = numvotes_by_runtime_sorted['numvotes'].astype(str)

# Remove commas from the 'numvotes' column
numvotes_by_runtime_sorted['numvotes'] = numvotes_by_runtime_sorted['numvotes'].str.replace(',', '')

# Convert the 'numvotes' column to numeric data type
numvotes_by_runtime_sorted['numvotes'] = pd.to_numeric(numvotes_by_runtime_sorted['numvotes'])

# Set the style and remove grid lines
sns.set_style("white")
sns.despine()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='runtime_interval', y='numvotes', data=numvotes_by_runtime_sorted, color='blue')
plt.title('Number of Votes by Runtime Interval')
plt.xlabel('Runtime Interval')
plt.ylabel('Number of Votes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Films revenue

##### Total Gross by Genre

The Adventure genre stands out prominently with the highest total gross, exceeding $114 billion, indicating its popularity and commercial success. Following closely behind, the Action genre also demonstrates significant revenue, surpassing $99 billion. Drama and Comedy genres exhibit substantial total gross, each exceeding $50 billion, showcasing their enduring appeal to audiences. On the other hand, Documentary genre has notably lower total gross, indicating its niche audience and limited commercial success compared to other genres

```python
genre_counts = movies_data4['genres'].str.split(',', expand=True).stack().value_counts()

# Select the top 10 genres
top_10_genres = genre_counts.head(10)
top_10_genres

# Fill NaN values in 'domestic_gross' and 'foreign_gross' columns with 0
movies_data4['domestic_gross'] = pd.to_numeric(movies_data4['domestic_gross'], errors='coerce').fillna(0)
movies_data4['foreign_gross'] = pd.to_numeric(movies_data4['foreign_gross'], errors='coerce').fillna(0)

# Create a new column 'total_gross' by adding 'domestic_gross' and 'foreign_gross'
movies_data4['total_gross'] = movies_data4['domestic_gross'] + movies_data4['foreign_gross']

# Drop rows with NaN values in the 'genres' column
movies_data4_cleaned = movies_data4.dropna(subset=['genres'])

# Calculate total gross for each genre in the top 10
total_gross_by_genre = {}
for genre in top_10_genres.index:
    total_gross_by_genre[genre] = movies_data4_cleaned[movies_data4_cleaned['genres'].str.contains(genre)]['total_gross'].sum()

# Convert the dictionary to a DataFrame
total_gross_by_genre_df = pd.DataFrame(list(total_gross_by_genre.items()), columns=['Genre', 'Total Gross'])

# Display the DataFrame
print(total_gross_by_genre_df)
```

```python
# Define colors for each genre
colors = ['blue', 'red', 'green', 'orange', 'purple', 'yellow', 'pink', 'cyan', 'lightcoral', 'lightblue']

# Sort the DataFrame by 'Total Gross' column in descending order
total_gross_by_genre_df_sorted = total_gross_by_genre_df.sort_values(by='Total Gross', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(total_gross_by_genre_df_sorted['Genre'], total_gross_by_genre_df_sorted['Total Gross'],color=colors)
plt.title('Total Gross by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Gross')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

##### Revenue per Film by Genre

The Adventure genre generates the highest revenue per film, averaging over $31 million per film, indicating its strong commercial performance and audience appeal. Action films also exhibit significant revenue per film, averaging around $15.8 million. Conversely, Documentary films have the lowest revenue per film, with an average of around $56,700, reflecting their niche audience and limited commercial potential compared to other genres. Drama, Horror, and Romance genres also demonstrate moderate revenue per film, while Comedy, Thriller, Crime, and Biography genres fall in between

```python
# Fill NaN values in 'domestic_gross' and 'foreign_gross' columns with 0
movies_data4['domestic_gross'] = pd.to_numeric(movies_data4['domestic_gross'], errors='coerce').fillna(0)
movies_data4['foreign_gross'] = pd.to_numeric(movies_data4['foreign_gross'], errors='coerce').fillna(0)

# Create a new column 'total_gross' by adding 'domestic_gross' and 'foreign_gross'
movies_data4['total_gross'] = movies_data4['domestic_gross'] + movies_data4['foreign_gross']

# Drop rows with NaN values in the 'genres' column
movies_data4_cleaned = movies_data4.dropna(subset=['genres'])

# Calculate total gross and number of films for each genre in the top 10
total_gross_by_genre = {}
num_films_by_genre = {}
for genre in top_10_genres.index:
    genre_movies = movies_data4_cleaned[movies_data4_cleaned['genres'].str.contains(genre)]
    total_gross_by_genre[genre] = genre_movies['total_gross'].sum()
    num_films_by_genre[genre] = len(genre_movies)

# Calculate revenue per film for each genre
revenue_per_film_by_genre = {genre: total_gross_by_genre[genre] / num_films_by_genre[genre] for genre in total_gross_by_genre}

# Convert the dictionary to a DataFrame
revenue_per_film_df = pd.DataFrame(list(revenue_per_film_by_genre.items()), columns=['Genre', 'Revenue per Film'])

# Sort DataFrame by 'Revenue per Film' in descending order
revenue_per_film_df = revenue_per_film_df.sort_values(by='Revenue per Film', ascending=False)
revenue_per_film_df

```

```python
# Define colors for each genre
colors = ['blue', 'red', 'green', 'orange', 'purple', 'yellow', 'pink', 'cyan', 'lightcoral', 'lightblue']

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(revenue_per_film_df['Genre'], revenue_per_film_df['Revenue per Film'], color=colors)
plt.title('Revenue per Film by Genre')
plt.xlabel('Genre')
plt.ylabel('Revenue per Film')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
